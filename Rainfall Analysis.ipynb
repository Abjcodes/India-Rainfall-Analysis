# Question no.2
Which months does it rain the most in India?

```python
plt.figure(figsize=(12,8))
df_months = df_in[['YEAR', 'Jan-Feb','Mar-May', 'Jun-Sep', 'Oct-Dec']].groupby('YEAR')
sns.lineplot(data = df_months.mean(),palette = "Set2", markers = True)
sns.set(style="white")
plt.xlabel("Year", fontsize=15)
plt.ylabel("Annual rainfall", fontsize=15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data = df_months.mean(), palette = "Set2")
sns.set(style="white")
plt.xlabel("Year", fontsize=15)
plt.ylabel("Annual rainfall", fontsize=15)
plt.grid()
plt.show()
```

The months between June to September recieves significantly higher rate of rain.
June-September is the monsoon season of India.
